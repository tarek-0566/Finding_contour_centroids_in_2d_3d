```python
import numpy as np

def line_intersection(line1, line2):
    """Calculates intersection point lines (if it exists).

    Args:
        line1: Line represented as a list or tuple: [(x0, y0), (x1, y1)]
        line2: Line represented similarly.

    Returns:
        intersection_point: A tuple (x, y) if intersection exists, otherwise None
    """
    (x0, y0), (x1, y1) = line1[0]
    (x2, y2), (x3, y3) = line2[0]

    dx1 = x1 - x0
    dy1 = y1 - y0
    dx2 = x3 - x2
    dy2 = y3 - y2

    det = dx1 * dy2 - dy1 * dx2
    if det == 0: 
        return None  # Lines are parallel or coincident

    t1 = ((x2 - x0) * dy2 + (y0 - y2) * dx2) / det 
    t2 = ((x2 - x0) * dy1 + (y0 - y2) * dx1) / -det  

    # Check if intersection lies within line segments
    if 0 <= t1 <= 1 and 0 <= t2 <= 1:
        intersection_x = x0 + t1 * dx1  
        intersection_y = y0 + t1 * dy1
        return intersection_x, intersection_y
    else:
        return None
```

```python
def calculate_intersection(p1, p2, p3, p4):
    """
    Calculates the intersection point of two lines using their slope-intercept form.

    Args:
        p1, p2: Coordinates of two points on the first line.
        p3, p4: Coordinates of two points on the second line.

    Returns:
        The intersection point (x, y) if they intersect, None otherwise.
    """

    # Get slopes and y-intercepts of both lines
    x1, y1 = p1
    x2, y2 = p2
    m1 = (y2 - y1) / (x2 - x1) if x2 - x1 != 0 else float('inf')  # Handle vertical line
    b1 = y1 - m1 * x1

    x3, y3 = p3
    x4, y4 = p4
    m2 = (y4 - y3) / (x4 - x3) if x4 - x3 != 0 else float('inf')  # Handle vertical line
    b2 = y3 - m2 * x3

    # Check for parallel or coincident lines
    if m1 == m2 and b1 == b2:
        return None  # Lines are coincident, infinite intersections

    elif m1 == m2 and b1 != b2:
        return None  # Lines are parallel, no intersection

    # Calculate intersection point
    if m1 != float('inf') and m2 != float('inf'):  # Both lines have finite slopes
        x = (b2 - b1) / (m1 - m2)
        y = m1 * x + b1
    elif m1 == float('inf'):  # First line is vertical
        x = x1
        y = m2 * x + b2
    else:  # Second line is vertical
        x = x3
        y = m1 * x + b1

    return (x, y)
```

```python
# Create empty lists to store points
line_points = []
intersections = []

# Extract points and calculate intersections (loop through detected lines)
for line in lines:
    x1, y1, x2, y2 = line[0]  # Access line endpoints from each line
    line_points.append([(x1, y1), (x2, y2)])  # Append as pairs to list

    # Calculate intersection with other lines (assuming 3 lines)
    if len(line_points) >= 4:
        for other_line in line_points[:2]:
            intersection = calculate_intersection(other_line[0], other_line[1], (x1, y1), (x2, y2))
            if intersection:
                intersections.append(intersection)
```

```python
def find_close_point_centroids(points, threshold):
    close_points = {}
    for i in range(len(points)):
        for j in range(i + 1, len(points)):
            distance = np.sqrt(((points[i][0] - points[j][0])**2) + ((points[i][1] - points[j][1])**2))
            if distance <= threshold:
                if i in close_points:
                    close_points[i].append(points[j])
                elif j in close_points:
                    close_points[j].append(points[i])
                else:
                    close_points[i] = [points[i], points[j]]

    centroids = {}
    for group_id, group in close_points.items():
        centroid_x = sum(point[0] for point in group) / len(group)
        centroid_y = sum(point[1] for point in group) / len(group)
        centroids[group_id] = (centroid_x, centroid_y)

    return centroids
```

```python
# Filter contours based on the number of coordinates
filtered_contours = [contour for contour in contours if len(contour) > 11]
```

```python
# Iterate through each contour
for contour in filtered_contours:
    # Calculate the centroid of the contour
    M = cv2.moments(contour)
    if M["m00"] != 0:
        cX = int(M["m10"] / M["m00"])
        cY = int(M["m01"] / M["m00"])
        print(cX, cY)
    else:
        cX, cY = 0, 0

    # Draw a circle at the centroid
    cv2.circle(image, (cX, cY), 3, (0, 0, 255), -1)
    # Put text with coordinates
    cv2.putText(image, f'({cX}, {cY})', (cX - 20, cY - 5), cv2.FONT_HERSHEY_SIMPLEX, 0.3, (255, 255, 255), 1)
```
