```python
#Find intersections of horizontal lines with contour edges
horizontal_intersections_1 =set()
for edge_start, edge_end in zip(rotated_contour[:-1], rotated_contour[1:]):
    x1, y1 = edge_start[0]
    x2, y2 = edge_end[0]
    for y_coord in horizontal_lines:
        if (y1 <= y_coord <= y2) or (y2 <= y_coord <= y1):
            if y2 - y1 != 0:  # Avoid division by zero
                x_intersection = int(x1 + (y_coord - y1) * (x2 - x1) / (y2 - y1))
                horizontal_intersections_1.add((x_intersection, y_coord))
            else:
                if y1 == y2:  # If both y coordinates are equal, consider it as an intersection
                    horizontal_intersections_1.add((x1, y1))

sorted_intersections_y = sorted(horizontal_intersections_1, key=lambda point: point[1])
print(sorted_intersections_y)
```

```python
# Draw intersection points
for point in horizontal_intersections_1:
    # Convert the floating-point values to integers
    point_int = (int(point[0]), int(point[1]))
    # Draw the circle at the intersection point
    cv2.circle(img_rotated, point_int, 2, 128, -1)

# Display the result
cv2_imshow(img_rotated)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
